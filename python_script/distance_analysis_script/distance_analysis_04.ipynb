```python
import numpy as np
import pandas as pd
```

```python
a = pd.read_table('/hdd/Taemin_before/chr_analysis/head/order_chr2-.txt', delim_whitespace=True, header = 'infer')
```

```python
b = pd.read_table('/hdd/Taemin_before/chr_analysis/head/order_chr2+.txt', delim_whitespace=True, header = 'infer')
```

```python
c = a.append(b)
```

```python
c['mean'] = c[['ExonEnd', 'ExonStart']].mean(axis=1)
```

```python
d = c[['name2', 'ExonStart', 'ExonEnd', 'mean']]
```

```python
e = d
```

```python
e['max'] = e[['ExonStart', 'ExonEnd']].apply(max, axis=1)
```

```python
e['min'] = e[['ExonStart', 'ExonEnd']].apply(min, axis=1)
```

```python
e[['diff']] = e[['max']].sub(e['min'], axis = 0).div(2)
```

```python
from IPython.display import clear_output
```

```python
temp_table = pd.DataFrame()
```

```python
t3 = pd.DataFrame()
for i in range(1501,2688):
    clear_output()
    t1 = e.iloc[i]
    for k in range(i, 2688):
        t2 = e.iloc[k]
        t3 = pd.DataFrame([[t1.name2, t2.name2, (t2['mean'] - t1['mean']) - (t1['diff'] + t2['diff'])]], columns=('from','to', 'distance'))
        temp_table = temp_table.append(t3)
        print(i, k)
```

```python
temp_table.to_csv('/hdd/Correlation_Disance/chr2_distance_04.csv')
```
