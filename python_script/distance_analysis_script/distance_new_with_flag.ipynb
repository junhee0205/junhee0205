```python
a['max'] = a[['ExonStart', 'ExonEnd']].apply(max, axis=1)
b['max'] = b[['ExonStart', 'ExonEnd']].apply(max, axis=1)

a['min'] = a[['ExonStart', 'ExonEnd']].apply(min, axis=1)
b['min'] = b[['ExonStart', 'ExonEnd']].apply(min, axis=1)

a['mean'] = a[['ExonEnd', 'ExonStart']].mean(axis=1)
b['mean'] = b[['ExonEnd', 'ExonStart']].mean(axis=1)
b[['diff']] = b[['max']].sub(b['min'], axis = 0).div(2)
a[['diff']] = a[['max']].sub(a['min'], axis = 0).div(2)
```

```python
from IPython.display import clear_output
```

```python
testing = pd.DataFrame()
for i in range(1, 600):
    clear_output()
    for k in range(1, b['name2'].count()):
        t1 = a.iloc[i]
        t2 = b.iloc[k]
        if t1['ExonEnd'] < t2['ExonStart']:
            flag = 'TT'
        elif t1['ExonStart'] < t2['ExonStart'] & t1['ExonEnd'] > t2['ExonStart'] & t1['ExonEnd'] < t2['ExonEnd']:
            flag = 'TT_fold'
        elif t1['ExonStart'] > t2['ExonEnd']:
            flag = 'HH'
        elif t1['ExonStart'] > t2['ExonStart'] & t1['ExonEnd'] < t2['ExonEnd'] & t1['ExonStart'] < t2['ExonEnd']:
            flag = 'HH_fold'
        elif t1['ExonStart'] > t2['ExonStart'] & t1['ExonEnd'] < t2['ExonEnd']:
            flag = 'minus_in_plus'
        elif t1['ExonStart'] < t2['ExonStart'] & t1['ExonEnd'] > t2['ExonEnd']:
            flag = 'plus_in_minus'
        elif t1['ExonStart'] < t2['ExonStart'] & t1['ExonEnd'] == t2['ExonStart']:
            flag = 'TT_touch'
        elif t1['ExonStart'] == t2['ExonEnd'] & t1['ExonEnd'] < t2['ExonEnd']:
            flag = 'HH_touch'
        t3 = pd.DataFrame([[t1.name2, t2.name2, abs(t2['mean'] - t1['mean']) - (t1['diff'] + t2['diff']), flag]], columns=('from','to', 'distance', 'flag'))
        testing = testing.append(t3)
        print(i, '/', a['name2'].count(), '|', k, '/', b['name2'].count())
```
